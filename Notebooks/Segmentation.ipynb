```python
import sys
sys.path.insert(1, "C:/Users/Azza/Desktop/Youssef/Mask-RCNN-TF2")
!pip install -r "C:/Users/Azza/Desktop/Youssef/Mask-RCNN-TF2/requirements.txt"
import mrcnn
import mrcnn.config
import mrcnn.model
import mrcnn.visualize
import cv2
import os

CLASS_NAMES = ['lesion']
class SimpleConfig(mrcnn.config.Config):
    # Give the configuration a recognizable name
    NAME = "Segmentation"
    
    # set the number of GPUs to use along with the number of images per GPU
    GPU_COUNT = 1
    IMAGES_PER_GPU = 1

	# Number of classes = number of classes + 1 (+1 for the background). The background class is named BG
    NUM_CLASSES = len(CLASS_NAMES) +1
    
    DETECTION_MIN_CONFIDENCE = 0.9

# Initialize the Mask R-CNN model for inference and then load the weights.
# This step builds the Keras model architecture.
model = mrcnn.model.MaskRCNN(mode="inference", 
                             config=SimpleConfig(),
                             model_dir=os.getcwd())

# Load the weights into the model.
model.load_weights(filepath="C:/Users/Azza/Desktop/Youssef/segmentationWeights.h5", 
                   by_name=True)

# load the input image, convert it from BGR to RGB channel
image = cv2.imread("C:/Users/Azza/Desktop/Youssef/ISIC 2019 Dataset/ISIC_2019_Training_Input/1/augment_ISIC_0000139_downsampled.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Perform a forward pass of the network to obtain the results
r = model.detect([image], verbose=0)

# Get the results for the first image.
r = r[0]

# Visualize the detected objects.
mrcnn.visualize.display_instances(image=image, 
                                  boxes=r['rois'], 
                                  masks=r['masks'], 
                                  class_ids=r['class_ids'], 
                                  class_names=CLASS_NAMES, 
                                  scores=r['scores'])
```
