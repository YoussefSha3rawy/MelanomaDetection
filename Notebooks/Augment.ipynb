```python
from tensorflow import keras
from imutils import paths
import os

from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest')

filenames = sorted(paths.list_images("C:/Users/Azza/Desktop/Youssef/ISIC 2019 Dataset/ISIC_2019_Training_Input/1"))

print(len(filenames))
for f in filenames:
    img = load_img(f)  
    x = img_to_array(img) 
   # Reshape the input image 
    x = x.reshape((1, ) + x.shape)  
    i = 0
    print(filenames.index(f)*100//len(filenames))
    prefix ="augment_"+ f.split('/')[-1][2:-4]+'#'
    for batch in datagen.flow(x, batch_size = 1,save_to_dir="C:/Users/Azza/Desktop/Youssef/ISIC 2019 Dataset/ISIC_2019_Training_Input/1",  
                      save_prefix =prefix, save_format ='jpeg'):
        i += 1
        if i > 0:
            break
```
